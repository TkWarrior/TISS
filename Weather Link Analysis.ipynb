```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('weather.csv',encoding='latin1')
```

```python
df.dropna(inplace=True)
df.dtypes
```

```python
df.shape
```

```python
df['Date & Time'] = pd.to_datetime(df['Date & Time'])
```

```python
df.dtypes
```

```python
df.set_index('Date & Time' , inplace=True)
```

```python

df.head()
```

```python
per_week = df[['Barometer - in Hg', 'High Bar - in Hg','Low Bar - in Hg']].resample('W')
```

```python
plt.plot(per_week['High Bar - in Hg'].max(),'.-', label = 'MAX')
plt.plot(per_week['Barometer - in Hg'].mean(),'.-', label = 'AVG')
plt.plot(per_week['Low Bar - in Hg'].min(),'.-', label = 'MIN')
plt.xticks(rotation=45)
plt.ylabel('Barometer - in Hg')
plt.title('Atmospheric pressure in Hg per week')
plt.grid(True)
plt.legend()
plt.gcf().autofmt_xdate()
```

```python
per_day = df[['Barometer - in Hg', 'High Bar - in Hg','Low Bar - in Hg']].resample('D')
```

```python
plt.plot(per_day['High Bar - in Hg'].max(),'.-', label = 'MAX')
plt.plot(per_day['Barometer - in Hg'].mean(),'.-', label = 'AVG')
plt.plot(per_day['Low Bar - in Hg'].min(),'.-', label = 'MIN')
plt.xticks(rotation=45)
plt.ylabel('Barometer - in Hg')
plt.title('Atmospheric pressure in Hg per day')
plt.grid(True)
plt.legend()
plt.gcf().autofmt_xdate()
```
